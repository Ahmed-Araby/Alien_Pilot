```python
import pickle 
import matplotlib.pyplot as plt 
```

```python
file = open('history.pickle', 'rb')
hist = pickle.load(file)
```

```python
print(hist)
```

```python
plt.plot(hist['angle_out_loss'])
plt.plot(hist['val_angle_out_loss'])

plt.title("angle loss, training vs validation ")
plt.xlabel("epoch number")
plt.ylabel("loss")

plt.legend(['train', 'validation'])
plt.savefig('angle_loss_train_vs_validation.png', dpi=300)

plt.show()
```

```python
plt.plot(hist['speed_out_loss'])
plt.plot(hist['val_speed_out_loss'])

plt.title("speed loss, training vs validation")
plt.xlabel("epoch number")
plt.ylabel("loss")

plt.legend(['train', 'validation'])
plt.savefig('speed_loss_train_vs_validation.png', dpi=300)

plt.show()
```
